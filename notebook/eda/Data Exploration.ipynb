```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
import natsort
```

```python
path_files = '/home/fahad/DATA/ML-project/ml-project/data/WorkingData/Normal_0/Normal_0_slices_DE/'
all_files = glob.glob(path_files+'DE_*.csv')  # path to slices
all_files_sorted = natsort.natsorted(all_files, reverse=False)
```

```python
list1= []
for filename in all_files_sorted:
    df = pd.read_csv(filename, index_col=None, header=0)
    df = df.drop('Unnamed: 0', axis=1)
    list1.append(df)

    
```

```python
num_slices = 20
%matplotlib inline
```

```python
for i in range(0, num_slices):
    plt.subplot(4, 5, i+1)
    plt.rcParams['figure.figsize'] = [30, 25]
    plt.plot(list1[i]['time'], list1[i]['vib'])
    plt.xlabel('time')
    plt.title('vibration')
    plt.ylim(-0.5, 0.5)
plt.show()
```

```python
fig1, (ax0, ax1) = plt.subplots(ncols=2)
plt.rcParams['figure.figsize'] = [30, 15]
ax0.plot(list1[0]['time'][0:100], list1[0]['vib'][0:100])
ax0.set_title('Drive End Data Normal')
ax0.set_xlabel('Time (t)')
ax0.set_ylabel('vibration ')
ax1.plot(list1[1]['time'][0:100], list1[1]['vib'][0:100])
ax1.set_title('Drive End Data Normal')
ax1.set_xlabel('Time (t)')
ax1.set_ylabel('vibration ')
```
